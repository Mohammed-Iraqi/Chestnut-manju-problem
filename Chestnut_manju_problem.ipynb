```python
# Assumptions:
# 1- The diameter of the solar system is 1,921.56 AU in diameter and AU (astronomical units) = 149,597,870.691 km. 
# 2- The initial size of the Chestnut has a diameter of 10 cm = 0.1 m = 0.0001 km
import numpy as np
t = 0 #Initial time
ri = 0.1/2 # initial radius
print("Initial Radious = {} meters".format(ri))
rf = 1921.56* 149.597870691E9/2 # final radius of the solar system
print("Final Solar System Radius = {} meters".format(rf))
Vi = (4/3)*np.pi*ri*ri*ri
print("Initial Volume = {} Cubic meters".format(Vi))
while True:
  Vi = 2*Vi
  ri = (3*Vi/4/np.pi)**(1/3)
  t = t + 300 # Time it takes to double the volume
  print("Time = {} Seconds".format(t))
  print("The current radius = {} meters".format(ri))
  if ri >= rf:
    break
    
print("Time for the Chestnut Buns to fill the solar system = {} Seconds".format(t))
print("Time for the Chestnut Buns to fill the solar system = {} minutes".format(t/60))
print("Time for the Chestnut Buns to fill the solar system = {} hours".format(t/60/60))
print("The final computed radius = {} meters".format(ri))
print("The Solar System Radius = {} meters".format(rf))




```
